## Packages

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
```

### Load data

```python
path = "./"
data = pd.read_csv(os.path.join(path,"OPTOS_Light_Images.csv"))
print(len(data))
data.head()
```

```python
data = data[['imageFilename','Class']]
data['imageFilename'] = data['imageFilename'].apply(lambda x: os.path.join(path,os.path.join('Light_Images',x)))
data['Class'] = data['Class'].apply(lambda x: int(x == "Good"))
data = data.sample(500,random_state = 42)
print(len(data))
data.head()
```

### Create network

```python
from Segmentation_OPTOS.GAN_OPTOS import CGAN_Fundus_vesion as cgan
```

```python
## Create the network
Net = cgan.CGAN(img_shape=(254,254,3))
```

## Train Network

```python
!mkdir images   ## Folder to store sampled images
Net.train(data,X = 'imageFilename',y = 'Class',epochs = 20000,batch_size = 32, sample_interval = 500)
```
